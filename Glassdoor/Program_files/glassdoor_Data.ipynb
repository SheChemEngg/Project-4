```python
import re
import math


# Define a function to convert string to numerical values
def string_to_numeric(value):
    if pd.notna(value):
        if isinstance(value, (int, float)):
            return value
        matches = re.findall(r'\d+', value)
        if matches:
            return int(matches[0]) * 10**6 if 'million' in value else int(matches[0]) * 10**9
    return None

# Apply the function to the 'Company_Revenue' column
glassdoor['Avg_Company_Revenue'] = glassdoor['Company_Revenue'].apply(string_to_numeric)

# Define a function for revenue average
def revenue_average(company_revenue):
    
    if pd.notna(company_revenue):       
        return (int(company_revenue) * (1.5))
    return None

# Create the 'Avg_Company_Revenue' column
glassdoor['Avg_Company_Revenue'] = glassdoor['Avg_Company_Revenue'].apply(revenue_average)
glassdoor.loc[glassdoor['Company_Revenue'].str.contains('Less'), 'Avg_Company_Revenue'] *= 0.5

glassdoor.head()
```

```python
import math

def size_average(company_size):
    
    if company_size == 'Unknown' or company_size == '-1':
        return int('0')
    
    else:
        # Split the string by space
        parts = company_size.split()
        
        if '+' in parts[0]:
            return int(parts[0].replace('+', ''))
        else:
            # Average of the two numbers
            average = int((int(parts[0]) + int(parts[2])) / 2)
            return average

# Apply the function to create the 'Avg_Company_Size' column
glassdoor['Avg_Company_Size'] = glassdoor['Company_Size'].apply(size_average)

glassdoor.head()
```

```python
# Drop 2 original columns

glassdoor.drop(['Company_Size', 'Company_Revenue'], axis=1, inplace=True)
glassdoor.info()
```

```python
# Filling missing 'Avg_Company_Revenue' with the most relevant value based on 'Avg_Company_Size'

size_revenue = {0: 1.5*(10**8), 25: 7.5*(10**5), 125: 1.5*(10**8), 350: 1.5*(10**9), 750: 1.5*(10**9), 3000: 1.5*(10**8), 7500: 3*(10**9), 10000: 1.5*(10**10)}

for index, row in glassdoor[glassdoor['Avg_Company_Revenue'].isna()].iterrows():
    for size, revenue in size_revenue.items():
        if row['Avg_Company_Size'] == size:
            glassdoor.at[index, 'Avg_Company_Revenue'] = revenue
glassdoor.head()
```

```python

def simplify_job_title(title):
    title = title.lower()
    if 'data' in title and ('scientist' in title or 'science' in title or 'analytics' in title):
        return 'Data Scientist'
    elif 'data' in title and 'engineer' in title:
        return 'Data Engineer'
    elif 'machine' in title and 'learning' in title:
        return 'Machine Learning Engineer'
    elif 'analyst' in title:
        return 'Data Analyst'
    elif 'director' in title:
        return 'Director'
    elif 'manager' in title:
        return 'Manager'
    else:
        return 'Research Scientist'

# Apply the function to the 'Job_Title' column
glassdoor['Job_Title'] = glassdoor['Job_Title'].apply(simplify_job_title)

glassdoor['Job_Title'].unique()
```

```python
# Change glassdoor['Competitors_y_n'] = '-1' to '0'
glassdoor['Competitors_y_n'] = glassdoor['Competitors_y_n'].replace('-1', '0')

# Change others to '1'
glassdoor.loc[glassdoor['Competitors_y_n']!='0', 'Competitors_y_n'] = '1' 

glassdoor['Competitors_y_n'] = glassdoor['Competitors_y_n'].astype(int)
glassdoor['Competitors_y_n'].head()
```

```python
# 5 Regions according to National Geographic

regions = {
    'West': "WA-MT-OR-ID-WY-CA-NV-UT-CO-AK-HI",
    'Southwest': "AZ-NM-TX-OK",
    'Midwest': "ND-SD-NE-KS-MN-IA-MO-WI-IL-MI-IN-OH",
    'Southeast': "KY-WV-VA-AR-TN-NC-SC-GA-AL-MS-LA-FL",
    'Northeast': "PA-NY-VT-ME-NH-MA-RI-CT-NJ-DE-MD-DC"
    }

# Function to map state to region
def state_to_region(state):
    for region, states in regions.items():
        if state in states.split('-'):
            return region
    # return None  # Return None if no region is found

# Create the 'job_region' column by applying the mapping function
glassdoor['job_region'] = glassdoor['job_state'].apply(state_to_region)

glassdoor['job_region'].head()
```

```python
for column in glassdoor.columns:
    glassdoor[column] = glassdoor[column].replace('-1', 'Unknown')
    glassdoor[column] = glassdoor[column].replace(-1, 0)

glassdoor.info()
```

```python
glassdoor.drop(['Year_Founded', 'Competitors_y_n', 'Seniority_y_n_na'], axis=1, inplace=True)
glassdoor.dtypes
```

```python
glassdoor = glassdoor[['job_state', 'job_region', 'Sector', 'Company_Age', 'Avg_Company_Revenue', 'Avg_Company_Size', 'Job_Title', 'Avg_Salary',\
                       'Min_Salary', 'Max_Salary', 'Usr_Rating', 'Python_y_n', 'R_y_n', 'Spark_y_n', 'AWS_y_n', 'Excel_y_n']]
glassdoor.info()
```
